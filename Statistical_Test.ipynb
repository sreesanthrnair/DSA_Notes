#1. What is Hypothesis Testing?

Hypothesis testing helps us decide whether there is enough evidence in a sample of data to support a certain claim about a population.

Null Hypothesis (H₀): Assumes no difference or no effect. (e.g., μ = μ₀)

Alternative Hypothesis (H₁): Assumes there is a difference or effect. (e.g., μ ≠ μ₀)

Significance level (α): The probability of rejecting H₀ when it is true (commonly 0.05 or 5%).

We calculate a test statistic (like Z or t) and compare it with a critical value, or compute a p-value to make the decision.

Sample size is large (n ≥ 30).

Data follows (or approximately follows) a normal distribution.

Types of Z-Test:

One-sample Z-test: Compare sample mean with population mean.

Two-sample Z-test: Compare means of two independent samples.

Z-test for proportions: Compare sample proportion with population proportion.

#4. P-Value

Definition: The probability of obtaining a test statistic equal to or more extreme than the one observed, under the null hypothesis.

Decision Rule:

If p ≤ α → Reject H₀ (evidence against H₀).

If p > α → Fail to reject H₀ (not enough evidence).

For example:

If α = 0.05 and p = 0.03 → Reject H₀.

If α = 0.05 and p = 0.10 → Fail to reject H₀.

#5. Steps to Perform Z or T-Test

State the hypotheses: H₀ and H₁.

Choose significance level α (e.g., 0.05).

Identify the test type (one-tail or two-tail).

Compute test statistic (Z or t).

Find the critical value from Z or t table OR calculate p-value.

Make a decision:

Reject H₀ if test statistic > critical value or p < α.

Otherwise, fail to reject H₀.

#6. Key Differences Between Z-Test and T-Test
Feature	Z-Test	T-Test
Population std (σ)	Known	Unknown
Sample size	Large (n ≥ 30)	Small (n < 30)
Distribution	Normal	t-distribution (heavier tails)
